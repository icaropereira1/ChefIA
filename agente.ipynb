O projeto contempla:

- **Dois agentes:** `analista_de_dados` e `estrategista_de_gestao`.
- **Duas tarefas** associadas a esses agentes: `analisa_performance_cardapio` e `gera_recomendacoes_proativas`.
- **Uma Crew** com processo sequencial (`insights_restaurante_crew`).

O objetivo Ã© simular o "Agente Proativo" que discutimos: o primeiro agente analisa os dados brutos e classifica o cardÃ¡pio, e o segundo agente usa essa anÃ¡lise para gerar recomendaÃ§Ãµes estratÃ©gicas em linguagem natural para o dono do restaurante.

```python
from dotenv import load_dotenv
_ = load_dotenv()
from crewai import Crew, Process, Agent, Task
```

```python
# --- Agente 1: O Analista de Dados (O CÃ©rebro) ---
analista_de_dados = Agent(
    role="Analista de Engenharia de CardÃ¡pio",
    goal="""Analisar dados brutos de vendas e fichas tÃ©cnicas para 
    calcular lucratividade e popularidade de cada item do cardÃ¡pio.""",
    backstory="""VocÃª Ã© um especialista em anÃ¡lise de dados (BI) focado no setor de A&B.
    Sua especialidade Ã© cruzar dados de vendas (popularidade) com dados de custo (lucratividade) 
    para classificar cada produto na Matriz de Engenharia de CardÃ¡pio 
    (Estrela, Burro de Carga, Quebra-cabeÃ§a, CÃ£o).""",
    verbose=True,
    allow_delegation=False
)
```

```python
# --- Agente 2: O Estrategista (O Consultor) ---
estrategista_de_gestao = Agent(
    role="Consultor e Estrategista de Restaurante",
    goal="""Transformar a anÃ¡lise tÃ©cnica de dados em recomendaÃ§Ãµes de negÃ³cio 
    prÃ¡ticas e acionÃ¡veis para o dono do restaurante.""",
    backstory="""VocÃª Ã© um consultor de gestÃ£o experiente, ex-dono de restaurante. 
    VocÃª entende a dor do gerente e seu trabalho Ã© traduzir a anÃ¡lise tÃ©cnica 
    (como 'Quebra-cabeÃ§a' ou 'Burro de Carga') em aÃ§Ãµes simples e diretas 
    que aumentam o lucro, como otimizar preÃ§os, sugerir promoÃ§Ãµes 
    ou criar posts para marketing.""",
    verbose=True,
    allow_delegation=False
)
```

```python
# --- Tarefa 1: Analisar os dados ---
# (Aqui, no 'description', vocÃª passaria os dados brutos como contexto)
analisa_performance_cardapio = Task(
    description="""Analise estes dados de vendas e custos de um restaurante. 
    Sua tarefa Ã© calcular a popularidade mÃ©dia e a lucratividade mÃ©dia, 
    e entÃ£o classificar cada produto em 'Estrela', 'Burro de Carga', 
    'Quebra-cabeÃ§a' ou 'CÃ£o'.

    **Dados para AnÃ¡lise:**
    [AQUI VOCÃŠ INSERIRIA OS DADOS, por exemplo, o conteÃºdo do seu DataFrame final]
    """,
    expected_output="""Um relatÃ³rio tÃ©cnico com a classificaÃ§Ã£o de cada item 
    e os valores mÃ©dios de popularidade e lucratividade encontrados.""",
    agent=analista_de_dados
)
```

```python
# --- Tarefa 2: Gerar as RecomendaÃ§Ãµes ---
gera_recomendacoes_proativas = Task(
    description="""Com base no relatÃ³rio de anÃ¡lise de cardÃ¡pio fornecido, 
    sua tarefa Ã© escrever 3 recomendaÃ§Ãµes proativas para o dono do restaurante.
    Foque nas oportunidades mais Ã³bvias (ex: o 'Quebra-cabeÃ§a' mais lucrativo 
    ou o 'Burro de Carga' mais popular).
    
    Escreva em um tom informal e direto, como se estivesse mandando um 
    WhatsApp para o gerente. Comece com 'E aÃ­, ZÃ©! Notei algumas coisas aqui:'""",
    expected_output="""Um texto curto e informal contendo 3 recomendaÃ§Ãµes acionÃ¡veis 
    para o gerente do restaurante.""",
    agent=estrategista_de_gestao
)
```

```python
# --- Montando a Crew ---
insights_restaurante_crew = Crew(
    agents=[analista_de_dados, estrategista_de_gestao],
    tasks=[analisa_performance_cardapio, gera_recomendacoes_proativas],
    process=Process.sequential,  # O Estrategista (Tarefa 2) depende da AnÃ¡lise (Tarefa 1)
    verbose=2
)
```

```python
# --- Executando o trabalho ---
# (VocÃª precisarÃ¡ alimentar os dados reais na Tarefa 1)
resultado = insights_restaurante_crew.kickoff()

print("\n\n--- RESULTADO FINAL DO AGENTE ---")
print(resultado)
```
